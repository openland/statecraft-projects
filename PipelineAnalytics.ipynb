```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

# Data Loading

```python
labels = [
    '2014 Q1',
    '2014 Q2',
    '2014 Q3',
    '2014 Q4',
    '2015 Q1',
    '2015 Q2',
    '2015 Q3',
    '2015 Q4',
    '2016 Q1',
    '2016 Q2',
    '2016 Q3',
    '2016 Q4',
    '2017 Q1'
]
completed = []
pipeline = []
for i in labels:
    completed.append(pd.read_csv('data/PipelineCompleted_' + i.replace(' ', '_') + '.csv', sep=',', infer_datetime_format=True))
    pipeline.append(pd.read_csv('data/PipelineCleaned_' + i.replace(' ', '_') + '.csv', sep=',', infer_datetime_format=True))
```

# Completed Projects with Net Units > 10

```python
for i in range(len(completed)):
    print(labels[i])
    display(completed[i][completed[i]['NET_UNITS'] > 10].sort_values('NET_UNITS', ascending=False))
```

# Statistics for each quarter

```python
labels_years = ['2014', '2015', '2016', '2017']
stats = pd.DataFrame(index=labels, columns=['NET_UNITS', 'NET_AFF_UNITS', 'AFF_SHARE', 'NET_UNITS_CONST', 'NET_AFF_CONST'])
stats_year = pd.DataFrame(index=labels_years, columns=['NET_UNITS', 'NET_AFF_UNITS', 'AFF_SHARE', 'NET_UNITS_CONST', 'NET_AFF_CONST'])
for i in range(len(completed)):
    net = completed[i]['NET_UNITS'].values.sum()
    net_aff = completed[i]['NET_AFF_UNITS'].values.sum()
    aff_share = round((net_aff / net) * 100, 2)
    
    pnet = pipeline[i]['NET_UNITS'].values.sum()
    pnet_aff = pipeline[i]['NET_AFF_UNITS'].values.sum()
    
    stats.set_value(labels[i], 'NET_UNITS', net)
    stats.set_value(labels[i], 'NET_AFF_UNITS', net_aff)
    stats.set_value(labels[i], 'AFF_SHARE', aff_share)
    stats.set_value(labels[i], 'NET_UNITS_CONST', pnet)
    stats.set_value(labels[i], 'NET_AFF_CONST', pnet_aff)
    
for i in range(len(labels_years)):
    net = 0
    net_aff = 0
    pnet = 0
    pnet_aff = 0
    for j in range(i * 4, min(i*4 +4, len(labels))):
        net += completed[j]['NET_UNITS'].values.sum()
        net_aff += completed[j]['NET_AFF_UNITS'].values.sum()
        pnet += pipeline[j]['NET_UNITS'].values.sum()
        pnet_aff += pipeline[j]['NET_AFF_UNITS'].values.sum()
    aff_share = round((net_aff / net) * 100, 2)
    
    stats_year.set_value(labels_years[i], 'NET_UNITS', net)
    stats_year.set_value(labels_years[i], 'NET_AFF_UNITS', net_aff)
    stats_year.set_value(labels_years[i], 'AFF_SHARE', aff_share)
    stats_year.set_value(labels_years[i], 'NET_UNITS_CONST', pnet)
    stats_year.set_value(labels_years[i], 'NET_AFF_CONST', pnet_aff)
    
display(stats)
display(stats_year)
```
